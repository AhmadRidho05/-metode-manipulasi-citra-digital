```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('img-3.jpg')
```

```python
if img is None:
    print("⚠️ Gambar tidak ditemukan! Pastikan file ada di folder yang sama.")
else:
    # OpenCV default BGR → ubah jadi RGB biar warnanya bener saat di-plot
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # ====== Brightness Adjustment ======
    # beta = nilai brightness (+ cerah, - gelap)
    brightness_value = 50  
    bright_img = cv2.convertScaleAbs(img_rgb, alpha=1, beta=brightness_value)
```

```python
# ====== Tampilkan Hasil ======
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img_rgb)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Brightness Adjusted")
plt.imshow(bright_img)
plt.axis("off")

plt.show()
```
